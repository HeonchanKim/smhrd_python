```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y = [2,4,6]
y
```

```python
plt.plot(y) # 그래프 그리기
plt.show() #그래프 출력
```

```python
y = [2,4,6]
x = [1,2,3]

plt.plot(x,y)
plt.show()
```

```python
# line style
# 그려지는 라인의 스타일을 지정하는 속성
#plt.plot(x,y,linestyle='--')
plt.plot(x,y,ls='--')
plt.show()
```

```python
x = [2,4,6,8]
y = [10,13,15,20]
y2 = [10,20,17,15]

plt.plot(x,y)
plt.plot(x,y2, ls=':')
plt.show()
```

```python
# marker
plt.plot(x,y, marker='o')
plt.show()
```

```python
# line width
# 라인 두께 지정하기 - lw
#plt.plot(x,y, linewidth='5')
plt.plot(x,y, lw='5')
plt.show()
```

```python
# line color
# 라인의 색상 변경하기 -> color
plt.plot(x,y,color='r')
plt.show()
# RGB 색상표를 이용할 수 있다.
```

```python
plt.plot(x,y,color='#5EC75E', marker='o', markerfacecolor='#0A6E0A')
plt.show()
```

```python
x = np.arange(7)
y = [1,4,5,8,9,5,3]
plt.plot(x,y,lw='5',ls='--',marker='o',mfc='r',ms='12',mec='#228B22',mew='4')
plt.show()
```

```python
# 그림의 영역 지정
# xlim : x 범위 지정
# ylim : y 범위 지정
# xlim(최소,최대)
plt.plot(x,y,lw='5',ls='--',marker='o',mfc='r',ms='15',mec='#228B22',mew='5')
plt.xlim(-2,8)
plt.ylim(-4,12)
plt.show()
```

```python
# grid
# 격자, 모눈종이 모양
plt.plot(x,y,lw='5',ls='--',marker='o',mfc='r',ms='12',mec='#228B22',mew='4')
plt.grid()
plt.show()
```

```python
# tick 설정
# xticks : x축 틱 설정
# yticks : y축 틱 설정
plt.plot(x,y,lw='5',ls='--',marker='o',mfc='r',ms='12',mec='#228B22',mew='4')
plt.xticks([0,3,6])
plt.yticks([1,5,9])
plt.grid()
plt.show()
```

```python
# 여러 그래프 동시에 그리기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, color='r')
plt.plot(y, color='k')
plt.plot(z, color='b')
plt.show()
```

```python
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x, color='r')
plt.show()
plt.plot(y, color='k')
plt.show()
plt.plot(z, color='b')
plt.show()
```

```python
# 범례 표시
plt.plot(x, label='x',)
plt.plot(y, label='y')
plt.plot(z, label='z')
plt.legend(loc='best')
plt.show()
# best
# upper right
# upper left
# lower left
# lower right
# right
# center left
# center right
# lower center
# upper center
# center
```

```python
plt.plot(x, label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')
plt.legend(loc='best')
plt.show()
```

```python
# 한글폰트 설정 방법
from matplotlib import rc
rc('font',family='Malgun Gothic')
```

```python
# 축에 타이틀 달기
# x 축 이름 : xlabel
# y 축 이름 : ylabel
# 그래프 이름 : title

plt.plot(x, label='엑스')
plt.plot(y, label='와이')
plt.plot(z, label='제트')
plt.xlabel('x축')
plt.ylabel('y축', rotation=0)
plt.title('x,y,z 그래프')
plt.legend(loc='best')
plt.show()
```

#### 요일별 사망 교통사고 시각화

```python
# '요일' 컬럼 가져오기
data['요일']
```

```python
# 중복 없는 데이터 출력
data.loc[:,'요일'].unique()
```

```python
# 요일별 사고 발생 건수 확인하기
temp=data['요일'].value_counts()
temp
```

```python
# 월 가져오기
temp['월']
```

```python
# 화 가져오기
temp['화']
```

```python
# 월, 화, 수 가져오기
temp[['월','화','수']]
```

```python
# 월 ~ 일 가져오기
y = temp[['월','화','수','목','금','토','일']]
y
```

```python
x = y.index
x
```

```python
# barchart 그리기
plt.bar(x,y)
plt.show()
```

```python
plt.bar(x,y)
plt.xlabel('요일')
plt.ylabel('사고 건수', rotation=0)
plt.title('요일별 사망교통사고 건수')
plt.show()
```

```python
# 특정 구간만 확대해서 보기
plt.bar(x,y)
plt.xlabel('요일')
plt.ylabel('사고 건수', rotation=0)
plt.title('요일별 사망교통사고 건수')
plt.ylim(500,620)
plt.show()
```
