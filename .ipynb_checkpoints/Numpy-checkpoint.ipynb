### BMI 구하기

```python
# ('파일명.확장자')
# delimiter : 구분 기호
data = np.loadtxt('height_weight.txt', delimiter=',')
```

```python
data
```

```python
#1. 데이터를 키와 몸무게로 나누기
height = data[0]
weight = data[1]
```

```python
#2. 키 단위 변경(cm -> m)
# 나누기 100 orr 곱하기 0.01
height_m = height*0.01
```

```python
height_m
```

```python
#3. BMI 연산
BMI = weight/(height_m**2)
```

```python
BMI
```

### Boolean 인덱싱
- Boolean : True/False
- True 값은 출력
- False 값은 출력하지 않음

```python
name = np.array(['세연','예진','예호','자연'])
name
```

```python
bol = np.array([True,False,True,False])
bol
```

```python
name[bol]
```

```python
# 데이터 생성
name_score = np.array([['세연',100],
            ['예진',90],
            ['예호',80],
            ['자연',70]])
```

```python
name_score
```

```python
#예진쌤의 점수?
name_score[1][1]
```

```python
# 예진쌤의 데이터
# 1. 이름 데이터만 인덱싱
name_score[:,0]
```

```python
# 2. 가져온 데이터에서 '예진' 인지 확인
name_score[:,0] == '예진'
# 불리언
```

```python
# 3. 조건에 따라 데이터 가져오기(인덱싱)
name_score[name_score[:,0] == '예진']
```

### 다양한 함수들

```python
# 1 ~ 10까지 담긴 1차원 배열 생성
arr = np.arange(1,11)
arr
```

```python
# 위에서 생성한 배열을 (2,5)크로 바꿔주기
arr = arr.reshape(2,5)
```

```python
arr
```

```python
# np.sum()
print('요소 전체 더하기 :',np.sum(arr))
print('axis=0 방향으로 더하기 :',np.sum(arr, axis=0))
print('axis=1 방향으로 더하기 :',np.sum(arr, axis=1))
```

```python
#np.mean()
print(np.mean(arr))
print(np.mean(arr,axis=0))
print(np.mean(arr,axis=1))
```
